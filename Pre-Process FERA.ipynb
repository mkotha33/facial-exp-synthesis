```python
import os
from PIL import Image
```

```python
base_path = '/home/harshitha/Desktop/dataset'
new_path = base_path + '_new1'
os.makedirs(new_path, exist_ok=True)
folders = os.listdir(base_path)
for x in folders:
    base_sub_path = os.path.join(base_path, x)
    new_base_sub_path = os.path.join(new_path, x)
    os.makedirs(new_base_sub_path, exist_ok=True)
    
    subs = os.listdir(base_sub_path)
    for y in subs:
        base_file_path = os.path.join(base_sub_path, y)
        new_base_file_path = os.path.join(new_base_sub_path, y)
        os.makedirs(new_base_file_path, exist_ok=True)
        
        files = os.listdir(base_file_path)
        ff_len = len(files)
        for i, ff in enumerate(files):
            if i%100 == 0:
                print(x,y,"Percentage Done:", round((i/ff_len)*100.0,ndigits=2))
            image_path = os.path.join(base_file_path, ff)
            new_image_path = os.path.join(new_base_file_path, ff)
            im = Image.open(image_path)
            resized_im = im.resize((256,256))
            resized_im.save(new_image_path)
            im.close()
print("Done!!")
```
